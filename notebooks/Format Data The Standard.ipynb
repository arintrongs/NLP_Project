```python
import pandas as pd
import numpy as np
from datetime import datetime
from pytz import timezone
```

```python
df = pd.read_csv('/Users/tomfoolery/Documents/Study/NLP_Project/data/the_standard_data(new).csv')
```

```python
df
```

```python
def applyview(x):
    if 'K' in x:
        temp = x[:-1:]
        temp = int(temp.replace(',',''))*1000
    else:
        temp = x
    return int(temp)
df['view'] = df['view'].apply(applyview)
```

```python
df.iloc[np.argmax(df['view'])]
```

```python
def addDOW(x):
    days = ['MON','TUE','WED','THU','FRI','SAT','SUN']
    return days[datetime.strptime(x,'%d.%m.%Y').weekday()]
```

```python
df['DOW'] = df['date'].apply(addDOW)
```

```python
def add543(x):
    temp = x.split('.')
    temp[-1] = str(int(temp[-1])+543)
    return '/'.join(temp)
```

```python
df['date'] = df['date'].apply(add543)
```

```python
df['time'] = df['time'].apply(lambda x : None)
df['like'] = df['like'].apply(lambda x : None)
df['comment'] = df['comment'].apply(lambda x : None)
df['share'] = df['share'].apply(lambda x : None)
```

```python
df.to_csv('the_standard_data.csv',index=False)
```
