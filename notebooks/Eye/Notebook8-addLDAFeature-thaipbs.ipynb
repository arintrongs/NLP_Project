```python
p_stemmer = PorterStemmer()
```

```python
def split_word(token):
    token = token.split(',')
#     print(token)
    stemmed_tokens = [p_stemmer.stem(i) for i in token]
    single_alpha_num_tokens = [i for i in stemmed_tokens if not i.isnumeric()]
    tokens = [i for i in single_alpha_num_tokens]
#     print(tokens)
    return tokens
```

```python
desc_list = []
for i in range(input_pbs.shape[0]):
    desc_list.append(input_pbs.iloc[i]['token'].split(','))
    

dictionary = corpora.Dictionary(desc_list)
dict2 = {dictionary[ID]:ID for ID in dictionary.keys()  }

corpus = [dictionary.doc2bow(text) for text in desc_list]
```

```python
# generate LDA model
num_top = 8
num_words = 8
num_it = 50
ldamodel = gensim.models.ldamodel.LdaModel(corpus,num_top, id2word = dictionary, random_state = 2, passes=num_it)
ldamodel.show_topics(num_top, num_words, log=True, formatted=False)
```

```python
# generate LDA model
num_top = 12
num_words = 8
num_it = 50
ldamodel2 = gensim.models.ldamodel.LdaModel(corpus,num_top, id2word = dictionary, random_state = 2, passes=num_it)
ldamodel2.show_topics(num_top, num_words, log=True, formatted=False)
```

```python
ldamodel = models.LdaModel.load('1 ' + str(num_top) + '_Topics_' + str(num_it) + '_Passes.model')
```

```python
def cosine_sim( list1, list2):
    #sim = 1 - spatial.distance.cosine(list1, list2)
    sumxx, sumxy, sumyy = 0, 0, 0
    for i in range(len(list1)):
        x = list1[i]; y = list2[i]
        sumxx += x*x
        sumyy += y*y
        sumxy += x*y
    sim = sumxy/math.sqrt(sumxx*sumyy)
    return sim

def similar(a, b):
    return abs(a - b) < 0.000001
    
def argmax(list):
        max = -1*float("inf")
        ind = 0
        for i in range(0, len(list)):
            if list[i] > max:
                ind = i
                max = list[i]
        return ind
```

```python
def list_to_string(ls):
    text = ""
    for element in ls:
        text+=str(element)
        text+=" "
    return text

def preprocess_project(proj_LDA,desc_list):

    proj_df = proj_LDA.copy()
    
    
    for i in range(num_top):
        proj_df['Topic '+str(i)] = 0.0

    proj_id_list = proj_df.index.unique()
   
    for i, text in enumerate(desc_list):
        bow = dictionary.doc2bow(text)
        topic_prob = ldamodel.get_document_topics(bow,minimum_probability=None, minimum_phi_value=None,per_word_topics=False)
        for j in range(len(topic_prob)):
            proj_df.at[proj_id_list[i], 'Topic '+str(topic_prob[j][0])] = topic_prob[j][1]
            
    return proj_id_list,proj_df


```

```python
proj_id_list,proj_LDA = preprocess_project(input_pbs,desc_list)
```
